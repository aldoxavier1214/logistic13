### Data splitting

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.25,stratify=df['CVD'],random_state=555)
```

### Feature engineering

```python
columnstobeimputed=['education', 'cigsPerDay', 'BPMeds','totChol', 'BMI', 'glucose']
```

```python
skew=['education', 'cigsPerDay', 'BPMeds','totChol', 'BMI', 'glucose']
for i in skew:
    print(df[i].skew())
```

```python
from sklearn.impute import SimpleImputer
si_col=SimpleImputer(strategy='median')
```

```python
train[columnstobeimputed]=si_col.fit_transform(train[columnstobeimputed])
```

```python
test[columnstobeimputed]=si_col.fit_transform(test[columnstobeimputed])
```

```python
test.isnull().sum()
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace=True)
```

```python
test.dropna(inplace=True)
```

### Model building

```python
X_train=train.drop(columns='CVD')
y_train=train[['CVD']]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

### Model score for train

```python
model.score(X_train,y_train)
```

```python
X_test=test.drop(columns='CVD')
y_test=test[['CVD']]
```

### Model score for test 

```python
model.score(X_test,y_test)
```

### Classification Report

```python
y_pred=model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

### ROC Curve

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
fpr,tpr,threshold=roc_curve(y_test,y_pred)
```

```python
roc_auc_score(y_test,y_pred)
```

```python
plt.figure(figsize=(5,3))
plt.plot(fpr,tpr,color='green')
plt.plot([0,1],[0,1],color='red')
plt.show()
```

## Stats model

```python
import statsmodels.api as sm
log_reg = sm.Logit(y_train, X_train).fit()
```

```python
print(log_reg.summary())
```

```python
ypred = log_reg.predict(X_test)
prediction = list(map(round, ypred))
print('Actual values', list(y_test.values))
print('Predictions :', prediction)
```

```python
from sklearn.metrics import (confusion_matrix, accuracy_score)
  
# confusion matrix
cm = confusion_matrix(y_test, prediction) 
print ("Confusion Matrix : \n", cm) 
  
# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, prediction))
```
